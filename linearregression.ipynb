```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Random X values
Y = 4 + 3 * X + np.random.rand(100, 1)  # Linear relation with noise
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, random_state=42)
```

```python
classifier = LinearRegression()
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
se = mean_squared_error(Y_test, Y_pred)
print("Squared Error:", se)
print("R^2 Score:", r2_score(Y_test, Y_pred))
```

```python
plt.scatter(X, Y, color='red', marker='x', label="Actual")
plt.plot(X_test, Y_pred, label='Prediction', color='blue')
plt.xlabel('X-data')
plt.ylabel('Y-data')
plt.legend()
plt.title('Linear Regression')
plt.grid()
plt.show()
```
